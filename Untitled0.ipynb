```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_hospital_layout():
    fig, ax = plt.subplots(figsize=(10, 12))
    ax.set_xlim(0, 40)
    ax.set_ylim(0, 30)
    ax.set_xticks(range(0, 41, 5))
    ax.set_yticks(range(0, 31, 5))
    ax.grid(True, linestyle='--', alpha=0.5)

    # Main areas
    ax.add_patch(patches.Rectangle((5, 20), 15, 8, edgecolor='black', facecolor='lightgray', lw=2, label='Reception & Waiting Area'))
    ax.text(7, 24, 'Reception & Waiting', fontsize=8, color='black')

    ax.add_patch(patches.Rectangle((5, 5), 10, 8, edgecolor='black', facecolor='lightgray', lw=2, label='Consultation Rooms'))
    ax.text(6, 9, 'Consultation Rooms', fontsize=8, color='black')

    ax.add_patch(patches.Rectangle((20, 5), 10, 8, edgecolor='black', facecolor='lightgray', lw=2, label='Pharmacy'))
    ax.text(22, 9, 'Pharmacy', fontsize=8, color='black')

    # Lift
    ax.add_patch(patches.Rectangle((18, 15), 4, 4, edgecolor='blue', facecolor='lightblue', lw=2, label='Lift'))
    ax.text(19, 16.5, 'Lift', fontsize=8, color='blue')

    # Staircase
    ax.add_patch(patches.Rectangle((24, 15), 6, 4, edgecolor='green', facecolor='lightgreen', lw=2, label='Staircase'))
    ax.text(25, 16.5, 'Staircase', fontsize=8, color='green')

    # Ramp
    ax.add_patch(patches.Polygon([[0, 15], [4, 18], [4, 12], [0, 15]], edgecolor='red', facecolor='pink', lw=2, label='Ramp'))
    ax.text(1, 14, 'Ramp', fontsize=8, color='red')

    # Main Entrance
    ax.add_patch(patches.Rectangle((15, 28), 10, 2, edgecolor='black', facecolor='white', lw=2, label='Main Entrance'))
    ax.text(17, 28.5, 'Main Entrance', fontsize=8, color='black')

    ax.set_title('Architectural Ground Floor Plan - Hospital')
    plt.show()

# Generate the layout
draw_hospital_layout()
```
